```
from time import sleep
#================ PRINT TOWERS ======================#

def print_tower(t1, t2, t3):
	global frame
	global max_rings

	labels=[]
	sy=10
	sx=20
	dy=20
	spaces=" "
	ring="#"
	spacer="  "*4
	
	t1=[0]*(max_rings-len(t1))+t1
	t2=[0]*(max_rings-len(t2))+t2
	t3=[0]*(max_rings-len(t3))+t3
	l=max(t1+t2+t3)
	

	for i in range(len(t1)):
		sp1=spaces*int((l-t1[i])/2)
		sp2=spaces*int((l-t2[i])/2)
		sp3=spaces*int((l-t3[i])/2)
		if t1[i] == 0:
			r1 = spaces
		else:
			r1 = t1[i]*"#"

		if t2[i] == 0:
			r2 = spaces
		else:
			r2 = t2[i]*"#"

		if t3[i] == 0:
			r3 = spaces
		else:
			r3 = t3[i]*"#"

		print(sp1+r1+sp1+spacer+sp2+r2+sp2+spacer+sp3+r3)
		



#================= TOWER OF HANOI MOVES ====================#
def move(s,d,t,n):
	global count


	if n == 1:
		d.insert(0,s.pop(0))
		#print("In single move: ",t1,t2,t3)
		count=count+1
		print_tower(t1, t2, t3)
		sleep(1)
		return 
	else:
		move(s,t,d,n-1)
		move(s,d,t,1)
		move(t,d,s,n-1)
	



# ============== MAIN PROGRAM ===================== #

max_rings=5

t1=[i for i in range(1,2*max_rings,2)]
t2=[]
t3=[]

count=0
print_tower(t1, t2, t3)
print("Starting config\n\n")
move(t1,t3,t2,max_rings)
print("Num Moves: ", count)

```
